# Outlier Treatment

checking outliers in variables

```python
#checking outliers with the help of boxplot

plt.boxplot(df);
```

there is many outliers in different varibles except Glucose and Outcome

```python
plt.boxplot(df["Insulin"]);
```

```python
x = np.quantile(df["Insulin"],[0,0.10,0.25,0.50,0.75,0.80,0.90,0.99])
x
```

```python
iqr = x[4]-x[2]
iqr
```

```python
uw = x[4]+1.5*iqr
uw
```

```python
lw = x[2]-1.5*iqr
lw
```

```python
df["Insulin"] = np.where(df["Insulin"]>uw,uw,df["Insulin"])
```

treating outliers

```python
def outlier_treatment():
    l = ["BMI","Glucose","SkinThickness","Age","BloodPressure","Insulin","Pregnancies","DiabetesPedigreeFunction"]
    for i in l:
        x = np.quantile(df[i],[0.25,0.75])
        iqr = x[1]-x[0]
        uw = x[1]+1.5*iqr
        lw = x[0]-1.5*iqr
        df[i]  = np.where(df[i]>uw,uw,(np.where(df[i]<lw,lw,df[i])))
        
outlier_treatment()
```

boxplot after treating outliers

```python
pno = 1
plt.figure(figsize = (15,10))
for i in ["BMI","Glucose","SkinThickness","Age","BloodPressure","Insulin","Pregnancies","DiabetesPedigreeFunction"]:
    if pno<9:
        plt.subplot(3,3,pno)
        plt.boxplot(df[i])
        plt.xlabel(i)
        pno+=1
```
